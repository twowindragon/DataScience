```python
sum(weather.avg_temp.isnull()), sum(weather.max_temp.isnull()), sum(weather.min_temp.isnull()), sum(weather.max_wind_speed.isnull()), sum(weather.most_wind_direction.isnull())
```

- 결측치 확인

```python
weather[weather.date == "2017-11-23"].min_temp.mean()
```

```python
weather.drop(weather[weather.most_wind_direction.isnull()].index, inplace = True)
```

- most_wind_direction의 널값들을 제거

```python
max_wind_speed_null_index = weather[weather.max_wind_speed.isnull()].index
max_wind_speed_null_date = weather[weather.max_wind_speed.isnull()].date.values 
max_wind_speed_null_index, max_wind_speed_null_date
```

```python
min_temp_null_index = weather[weather.min_temp.isnull()].index
min_temp_null_date = weather[weather.min_temp.isnull()].date.values 
min_temp_null_index, min_temp_null_date
```

```python
max_temp_null_index = weather[weather.max_temp.isnull()].index
max_temp_null_date = weather[weather.max_temp.isnull()].date.values
max_temp_null_index, max_temp_null_date
```

```python

for idx, date in enumerate(max_wind_speed_null_date):
     weather.loc[max_wind_speed_null_index[idx], "max_wind_speed"] = weather[date == weather.date].max_wind_speed.mean()

for idx, date in enumerate(min_temp_null_date):
     weather.loc[min_temp_null_index[idx], "min_temp"] = weather[date == weather.date].min_temp.mean()

for idx, date in enumerate(max_temp_null_date):
     weather.loc[max_temp_null_index[idx], "max_temp"] = weather[date == weather.date].max_temp.mean()
```

 - 널 값과 같은 날짜의 모든 지역의 평균으로 널값들을 채워줌

```python
weather.avg_temp = weather.avg_temp.fillna(value=(weather.min_temp + weather.max_temp) / 2)
```

- 평균 온도의 경우 최고온도와 최저온도의 평균으로 널값을 채움

```python
sum(weather.avg_temp.isnull()), sum(weather.max_temp.isnull()), sum(weather.min_temp.isnull()), sum(weather.max_wind_speed.isnull()), sum(weather.most_wind_direction.isnull())
```

```python
month = ["01", "02", "03", "04", "05", "06", "07", "08", "09", "10", "11", "12"]
col = weather.province.unique()
month_avg_temp = [[] for _ in range(16)]
for idx, province in enumerate(weather.code.unique()):
    
    province_weather = weather[weather.code == province]
    province_weather.loc[:]['date'] = province_weather[:]['date'].str[5:7]
    

    for m in month:
        month_avg_temp[idx].append(province_weather[province_weather.date == m].avg_temp.mean())

df4 = pd.DataFrame(month_avg_temp, index=col, columns= range(1, 13))

df4 = df4.T
df4.reset_index(inplace=True)
df4
```
